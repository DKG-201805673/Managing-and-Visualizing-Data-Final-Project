## Dataset

The dataset that I am using for this report comes from the COVID-19 Data Repository by CSSE (Center for Systems Science and Engineering) at the Johns Hopkins University from the Google BigQuery public datasets. The data includes the location and number of confirmed COVID-19 cases, deaths and recoveries for all affected countries. It was developed to enable researchers, public health authorities and the public to track the outbreak.

After reviewing the dataset on the Google BigQuery website I saw that it was too big of a dataset to extract in its entirety, and so I chose certain columns of data that I wanted to use for my report. I decided to use the data about the countries, confirmed cases, deaths and recovered cases just from the months of 2020. I also limited my dataset to only the top 10 countries that had the highest number of confirmed cases since the dataset contained data for all countries and I thought that restricting the data I used to the most affected countries would be better to analysis and report on.

I extracted the dataset using an API key and SQL query as shown below and then saved the dataframe created to a CSV file that I use throughout the rest of the report. 

```python
from google.cloud import bigquery
client = bigquery.Client()
#Querying the dataset from Google BigQuery using SQL
query_job = """
SELECT 
extract(MONTH from date) as month, country_region,
SUM(confirmed) as confirmed,
SUM(deaths) as deaths,
SUM(recovered) as recovered,
from `bigquery-public-data.covid19_jhu_csse.summary`
where extract(YEAR from date) = 2020
and country_region in ('US', 'India','Brazil','United Kingdom','Russia','France','Turkey','Spain','Italy','Germany')
group by month, country_region
order by country_region, month asc
"""
#Saving the dataframe to a CSV file
query = client.query(query_job).to_dataframe()
query.to_csv('ST445 Dataset.csv', index=False)
```

The last analysis I am going to do is the clustering of the COVID-19 data for confirmed cases against deaths. For this I will be using the whole dataset, not just the top 10 countries as it will give a better view of the relationship between the two variables.

```python
from google.cloud import bigquery
client = bigquery.Client()

query2_job = """
SELECT 
country_region,
SUM(confirmed) as confirmed,
SUM(deaths) as deaths
from `bigquery-public-data.covid19_jhu_csse.summary`
where extract(YEAR from date) = 2020
group by country_region
order by country_region
"""

query2 = client.query(query2_job).to_dataframe()
query2.to_csv('Cluster.csv', index=False)
```

```python
from sklearn.cluster import KMeans

df0 = pd.read_csv('Cluster.csv')

#print (df0.isnull().sum())

# remove nulls
df1 = df0.dropna()

#print (df1.isnull().sum())

#print (df1.shape)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df1['confirmed'],df1['deaths'])
current_xvalues = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_xvalues])
current_yvalues = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_yvalues])
plt.xlabel("Confirmed")
plt.ylabel("Deaths")

km = KMeans(n_clusters=4)

xfeatures = df1[['confirmed','deaths']]

clusters = km.fit_predict(xfeatures)

#print(km.labels_)

print(km.cluster_centers_)
```

```python
df1['cluster_components'] = clusters

plt.figure(figsize=(12, 6))
plt.scatter(df1['confirmed'],df1['deaths'],c=df1['cluster_components'],cmap='rainbow')
current_xvalues = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_xvalues])
current_yvalues = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_yvalues])
plt.xlabel("Confirmed")
plt.ylabel("Deaths")

print(df1[df1['cluster_components']==0])

print(df1[df1['cluster_components']==1])

print(df1[df1['cluster_components']==2])

print(df1[df1['cluster_components']==3])
```

#### Analysis of scatterplot clusters on COVID-19 data
From the scatterplot, the clusters all follow a linear pattern reflecting a positive correlation between the number of confirmed cases and the number of deaths attributable to COVID-19. The majority of countries are concentrated in a cluster towards the bottom left of the graph showing that most countries tried to limit the number of cases by implementing lockdown policies to suppress the spread of the virus which had a direct impact on the number of attributable deaths. The red cluster identifies countries with a higher death per confirmed cases indicating a higher level of infection which also resulted in a higher level of deaths. The blue and yellow clusters in the middle and top right of the graph represent the top 3 countries (Brazil, India, and the US) that had the highest number of confirmed cases and deaths. This scatter plot gives a good indication of how countries that more confirmed cases had generally also had more deaths reported. This could suggest that those countries did more testing of COVID-19 and were able to identify more accurately those that had died from COVID-19. In the clusters closer to the bottom left of the graph, those countries may not have tested as much, hence did not record as many deaths that were associated with COVID-19 infections.
